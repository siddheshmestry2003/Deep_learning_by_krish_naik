```python
# import necessary libary
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import pickle
```

```python
# load the data set
data=pd.read_csv("Churn_Modelling.csv")
data.head()
```

```python
# preprosess the data 
# drop irrelevent feature
data1=data.drop(['RowNumber','CustomerId','Surname'],axis=1)
data1
```

```python
# encode categorical variable
# this transform gender into number like male=1 or female=0

label_encoder_gender=LabelEncoder()
data1["Gender"]=label_encoder_gender.fit_transform(data1['Gender'])
data1
```

```python
# to convert geographical data into vector we use  one hot encoder
# OneHot ENCODER
from sklearn.preprocessing import OneHotEncoder
onehot_encoder_geography=OneHotEncoder()
geo_encoded_value=onehot_encoder_geography.fit_transform(data1[['Geography']])
geo_encoded_value
```

```python
onehot_encoder_geography.get_feature_names_out(['Geography'])
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# Step 1: One-hot encode
onehot_encoder_geography = OneHotEncoder(sparse_output=False)  # use sparse=False if older sklearn
geo_encoded_value = onehot_encoder_geography.fit_transform(data1[['Geography']])

# Step 2: Convert encoded matrix to DataFrame with proper column names
geo_encoded_df = pd.DataFrame(
    geo_encoded_value,
    columns=onehot_encoder_geography.get_feature_names_out(['Geography'])
)

# Step 3: Reset index before combining (to avoid misalignment)
geo_encoded_df.index = data1.index

# Step 4: Combine with the original DataFrame
data1 = pd.concat([data1.drop('Geography', axis=1, errors='ignore'), geo_encoded_df], axis=1)
```

```python
# sanve the encoder and scalar
with open('label_encoder_gender.pkl','wb') as file:
    pickle.dump(label_encoder_gender,file)

with open('oneHot_encoder_geo.pkl','wb') as file:
    pickle.dump(onehot_encoder_geography,file)
```

```python
# divide data set into indipendent and dependent feature
data1.head()
```

```python
x=data1.drop('Exited',axis=1)
y=data1['Exited']

#split the data in trainig and testing set
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

# scale this feature
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

# ANN IMPLEMENTATION USING TENSORFLOW AND KERAS

```python
## what i use 
# sequential network
# dense  ->64
# activation function
# optimizer -> backpropagation -> update the weight
# loss  [matrix are (accuracy,mse,mae)]

# training ->logs in folder -> using tensorboard -> visualization
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
x_test.shape[1]
```

```python
(x_train.shape[1],)
```

```python
# build our ANN Model
model=Sequential([
    Dense(64,activation='relu',input_shape=(x_train.shape[1],)) ,# first hidden layer connected with input layer
    Dense(32,activation='relu') , # second hidden layer
    Dense(1, activation='sigmoid')  #output layer


])
```

```python
import tensorflow
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.01)
loss=tensorflow.keras.losses.BinaryCrossentropy()
loss
```

```python
model.summary()
```

```python
## compile the model
model.compile(optimizer=opt,loss="binary_crossentropy",metrics=['accuracy'])
```

```python
# set up the tensorboard
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard

log_dir="logs/fit"+ datetime.datetime.now().strftime("%Y%m%d - %H%M%S")

tensorflow_callback=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
# set up Early Stoping
EarlyStopping_callback=EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)
```

```python
## training the model

history=model.fit(
    x_train,y_train,validation_data=(x_test,y_test),epochs=100,
    callbacks=[tensorflow_callback,EarlyStopping_callback]

)
```

```python
model.save('model.h5') # h5 file is comfortablewith keras
```
