```python
# import necessary libary
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import pickle
```

```python
# load the data set
data=pd.read_csv("Churn_Modelling.csv")
data.head()
```

```python
# preprosess the data 
# drop irrelevent feature
data1=data.drop(['RowNumber','CustomerId','Surname'],axis=1)
data1
```

```python
# encode categorical variable
# this transform gender into number like male=1 or female=0

label_encoder_gender=LabelEncoder()
data1["Gender"]=label_encoder_gender.fit_transform(data1['Gender'])
data1
```

```python
# to convert geographical data into vector we use  one hot encoder
# OneHot ENCODER
from sklearn.preprocessing import OneHotEncoder
onehot_encoder_geography=OneHotEncoder()
geo_encoded_value=onehot_encoder_geography.fit_transform(data1[['Geography']])
geo_encoded_value
```

```python
onehot_encoder_geography.get_feature_names_out(['Geography'])
```

```python
# combine one ht encoder columns with the orginal data 
data1 = pd.concat([data1.drop('Geography', axis=1, errors='ignore'), geo_encoded_value], axis=1)
```
